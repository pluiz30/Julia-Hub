## Resumo

Essa aula ilustrará dois dos mais importantes teoremas da probabilidade e da Estatística: a Lei dos Grandes Números (LGN) e o Teorema do Limite Central (TLC).

Estes lindos teoremas estão por trás de muitos resultados fundamentais em Econometria e modelagem econômica.

A aula é baseada em aplicações do TLC e da LGN.

Nós também demonstramos como a LGN e o TLC desagregam-se quando as suposições em que se baseiam não se mantêm.

Em adição, nós examinamos várias extensões úteis dos teoremas clássicos, tal como:

- O método delta, para funções suaves de variáveis aleatórias, e  
- O caso multivariado.  


Algumas dessas extensões são apresentadas nos exercícios.

## Relação

O TLC refina a LGN.

A LGN fornece condições sob as quais os momentos da amostra convergem para os momentos da população à medida que o tamanho da amostra aumenta.

O TCL provê informações sobre a taxa na qual os momentos da amostra convergem para os momentos da população à medida que o tamanho da amostra aumenta.

## Lei dos Grandes Números


<a id='index-4'></a>
Nós começaremos com a lei dos grandes números, a qual nos diz que médias simples irão convergir para médias populacionais.


<a id='lln-ksl'></a>

### A Lei dos Grandes Números clássica

A Lei dos Grandes Números clássica preocupa-se com variáveis aleatórias identicamente independentes (IID).

Aqui está a versão mais forte da LGN clássica, conhecida como *lei forte de Kolmogorov*.

Sejam $ X_1, \ldots, X_n $ variáveis aleatórias independentes e identicamente distribuídas, com distribuição comum $ F $.

Quando isso existir, seja $ \mu $ a média comum desta amostra:

$$
\mu := \mathbb E X = \int x F(dx)
$$

Em adição, seja

$$
\bar X_n := \frac{1}{n} \sum_{i=1}^n X_i
$$

A lei forte de Kolmogorov afirma que, se $ \mathbb E |X| $ é finita, então


<a id='equation-lln-as'></a>
$$
\mathbb P \left\{ \bar X_n \to \mu \text{ as } n \to \infty \right\} = 1 \tag{1}
$$

Qual o significado dessa última expressão?

Vamos pensar sobre a perspectiva da situação, imaginando por um momento que nosso computados pode gerar amostras aleatórias perfeitas (claro que [não pode](https://en.wikipedia.org/wiki/Pseudorandom_number_generator)).

Vamos também imaginar que ppodemos gerar sequências infinitas, para que a afirmação $ \bar X_n \to \mu $ tenha valor.

Nessa configuração, [(1)](#equation-lln-as) deve ser interpretado no sentido de que a probabilidade do computador produzir uma sequência onde $ \bar X_n \to \mu $ falha, é zero.


### Prova


<a id='index-5'></a>
A prova da lei forte de Kolmogorov não é trivial – veja, por exemplo, o teorema 8.3.5 de [[Dud02] (https://lectures.quantecon.org/zreferences.html#dudley2002).

Por outro lado, nós podemos provar uma versão fraca da LGN muito facilmente e ainda pegar a maior parte da intuição.

Nós provamos a versão como se segue: 

Se $ X_1, \ldots, X_n $ é IID com $ \mathbb E X_i^2 < \infty $, então, para qualquer $ \epsilon > 0 $, nós temos:


<a id='equation-lln-ip'></a>
$$
\mathbb P \left\{ | \bar X_n - \mu | \geq \epsilon \right\} \to 0
\quad \text{as} \quad
n \to \infty \tag{2}
$$

(Essa versão é fraca porquê nós clamamos apenas a [convergência na probabilidade](https://en.wikipedia.org/wiki/Convergence_of_random_variables#Convergence_in_probability) ao invés da [convergência quase certa](https://en.wikipedia.org/wiki/Convergence_of_random_variables#Almost_sure_convergence), e assume um segundo momento finito).

Para ver que é realmente isso, fixe $ \epsilon > 0 $, e seja $ \sigma^2 $ a variância de cada $ X_i $.

Relembre-se da [Desigualdade de Chebyshev](https://en.wikipedia.org/wiki/Chebyshev%27s_inequality), que nos diz que:


<a id='equation-lln-cheb'></a>
$$
\mathbb P \left\{ | \bar X_n - \mu | \geq \epsilon \right\}
\leq \frac{\mathbb E [ (\bar X_n - \mu)^2]}{\epsilon^2} \tag{3}
$$

Agora observe:

$$
\begin{aligned}
    \mathbb E [ (\bar X_n - \mu)^2 ]
    & = \mathbb E \left\{ \left[
    \frac{1}{n} \sum_{i=1}^n (X_i - \mu)
    \right]^2 \right\}
    \\
    & = \frac{1}{n^2} \sum_{i=1}^n \sum_{j=1}^n \mathbb E (X_i - \mu)(X_j - \mu) \nonumber
    \\
    & = \frac{1}{n^2} \sum_{i=1}^n \mathbb E (X_i - \mu)^2 \nonumber
    \\
    & = \frac{\sigma^2}{n} \nonumber
\end{aligned}
$$

Aqui está o passo crucial da terceira equação, que segue da independência.

Independência significa que, se $ i \not= j $, então o termo de covariância $ \mathbb E (X_i - \mu)(X_j - \mu) $ cai fora.

Como resultado, $ n^2 - n $ termos desaparecem, nos levando à uma última expressão que vai para zero em $ n $.

Combinando nosso último resultado com [(3)](#equation-lln-cheb), chegamos a estimar:


<a id='equation-lln-cheb2'></a>
$$
\mathbb P \left\{ | \bar X_n - \mu | \geq \epsilon \right\}
\leq \frac{\sigma^2}{n \epsilon^2} \tag{4}
$$

A pretensão em [(2)](#equation-lln-ip) agora é clara.

É claro, se a sequência $ X_1, \ldots, X_n $ é correlacionada, então os termos do produto cruzado $ \mathbb E (X_i - \mu)(X_j - \mu) $ não são necessariamente zero.

Enquanto isso não significar que a mesma linha de argumento é impossível, isso significa que se quisermos um resultado similar,  então as covariâncias serão “quase zero” para a “maioria” desses termos.

Numa longa sequência, isso pode ser verdade se, por exemplo, $ \mathbb E (X_i - \mu)(X_j - \mu) $ se aproximou de zero quando a diferença entre $ i $ e $ j $ se tornou grande.

Em outras palavras, a LGN pode trabalhar ainda se a sequência $ X_1, \ldots, X_n $ tiver um tipo de “independência assintótica”, no sentido de que a correlação vai a zero à medida que as variáveis se afastam da sequência.

Essa idéia é muito importante em análise de séries temporais, e nós iremos voltar nesse assunto em breve.

```julia
function ksl(distribution, n = 100)
    title = nameof(typeof(distribution))
    observations = rand(distribution, n)
    sample_means = cumsum(observations) ./ (1:n)
    μ = mean(distribution)
    plot(repeat((1:n)', 2),
         [zeros(1, n); observations'], label = "", color = :grey, alpha = 0.5)
    plot!(1:n, observations, color = :grey, markershape = :circle,
          alpha = 0.5, label = "", linewidth = 0)
    if !isnan(μ)
        hline!([μ], color = :black, linewidth = 1.5, linestyle = :dash, grid = false,
               label = ["Mean"])
    end
    plot!(1:n, sample_means, linewidth = 3, alpha = 0.6, color = :green,
          label = "Sample mean")
    return plot!(title = title)
end
```

### Média infinita

O que acontece se a condição $ \mathbb E | X | < \infty $ na afirmação da LGN não é satisfeita?

Esse pode ser o caso de a distribuição subjacente ter a cauda atada — o mais conhecido exemplo é a distribuição de Cauchy, cuja densidade é

$$
f(x) = \frac{1}{\pi (1 + x^2)} \qquad (x \in \mathbb R)
$$

A próxima figura mostra 100 desenhos independentes dessa distribuição.

```julia
Random.seed!(0); # resultados reproduzíveis
```

```julia
ksl(Cauchy())
```

Perceba o quanto as observações extremas prevalecem bem mais aqui do que na figura anterior.

Vamos agora dar uma olhada no comportamento da média simples:

```julia
Random.seed!(0); # resultados reproduziveis
```

```julia
function plot_means(n = 1000)
    sample_mean = cumsum(rand(Cauchy(), n)) ./ (1:n)
    plot(1:n, sample_mean, color = :red, alpha = 0.6, label = "Sample Mean", linewidth = 3)
    return hline!([0], color = :black, linestyle = :dash, label = "", grid = false)
end

plot_means()
```

Aqui temos aumentado $ n $ para 1000, mas a sequência ainda não mostra sinal de convergência.

A convergência estará visível se obtivermos um $ n $ ainda maior?

A resposta é: não.

Para ver isso, lembre-se que a [função característica](https://en.wikipedia.org/wiki/Characteristic_function_%28probability_theory%29) da Distribuição de Cauchy é:


<a id='equation-lln-cch'></a>
$$
\phi(t) = \mathbb E e^{itX} = \int e^{i t x} f(x) dx = e^{-|t|} \tag{5}
$$

Usando independência, a função característica da média simples se torna:

$$
\begin{aligned}
    \mathbb E e^{i t \bar X_n }
    & = \mathbb E \exp \left\{ i \frac{t}{n} \sum_{j=1}^n X_j \right\}
    \\
    & = \mathbb E \prod_{j=1}^n \exp \left\{ i \frac{t}{n} X_j \right\}
    \\
    & = \prod_{j=1}^n \mathbb E \exp \left\{ i \frac{t}{n} X_j \right\}
    = [\phi(t/n)]^n
\end{aligned}
$$

Segundo [(5)](#equation-lln-cch), isso é apenas $ e^{-|t|} $.

Logo, no caso da distribuição de Cauchy, a média simples por si tem a mesma distribuição de Cauchy, independentemente de $ n $.

Em particular, a sequência $ \bar X_n $ não converge para um ponto.

## Teorema do Limite Central


<a id='index-7'></a>
Agora, voltaremos para o Teorema do Limite Central, que nos fala sobre a distribuição dos desvios das médias amostral e populacional.

### Afirmação do teorema

O teorema do limite central é um dos resultados mais notáveis em toda a matemática.

Na configuração iid clássica, nos diz o seguinte:


<a id='statement-clt'></a>
Se a sequência $ X_1, \ldots, X_n $ é IID, com média
$ \mu $ e variância $ \sigma^2 \in (0, \infty) $, então


<a id='equation-lln-clt'></a>
$$
\sqrt{n} ( \bar X_n - \mu ) \stackrel { d } {\to} N(0, \sigma^2)
\quad \text{as} \quad
n \to \infty \tag{6}
$$

Aqui $ \stackrel { d } {\to} N(0, \sigma^2) $ indica [convergência na distribuição](https://en.wikipedia.org/wiki/Convergence_of_random_variables#Convergence_in_distribution) para uma normal centrada (isto é, com média zero) com desvio-padrão $ \sigma $.

### Intuição


<a id='index-8'></a>
A implicação direta do TLC é que para **qualquer** distribuição com o segundo momento finito, a simples operação de adicionar cópias independentes **sempre** leva a uma curva gaussiana.

Uma prova relativamente simples do teorema do limite central pode ser obtida via trabalho com funções características (veja, por ex., o teorema 9.5.6 de [Dud02](https://lectures.quantecon.org/zreferences.html#dudley2002)).

A prova é elegante mas desgastante, e ainda provê intuições simples e independentes.

De fato, todas as provas do TLC que nós conhecemos são similares nesse ponto.

Por quê a adição de cópias independentes produz uma distribuição em forma de sino?

Parte da resposta pode ser obtida pela investigação adicional de variáveis aleatórias independentes de Bernoulli.

Em particular, seja $ X_i $ binária, com $ \mathbb P\{X_i = 0\} = \mathbb P\{X_i = 1 \} = 0.5 $, e seja $ X_1, \ldots, X_n $ independente.

Pense $ X_i = 1 $ como um  “sucesso”, então $ Y_n = \sum_{i=1}^n X_i $ é o número de sucessos em $ n $ lançamentos.

A próxima figura plota a função de probabilidade de $ Y_n $ para $ n = 1, 2, 4, 8 $:

```julia
binomial_pdf(n) =
    bar(0:n, pdf.(Binomial(n), 0:n),
        xticks = 0:10, ylim = (0, 1), yticks = 0:0.1:1,
        label = "Binomial($n, 0.5)", legend = :topleft)
```

```julia
plot(binomial_pdf.((1,2,4,8))...)
```

Quando $ n = 1 $, a distribuição é dobrada — sucesso e não-sucesso tem a mesma probabilidade.

Quando $ n = 2 $ podemos ainda ter 0, 1 ou 2 sucessos.

Observe que o pico no grosso da probabilidade está no ponto médio $ k=1 $.

A razão é que ali há mais caminhos para se obter 1 sucesso (“falhar e ter sucesso” ou “ter sucesso e falhar”) do que obter zero ou dois sucessos.

Além disso, os dois experimentos são independentes, então as oportunidades “falhar e ter sucesso” e “ter sucesso e falhar” são exatamente como as oportunidades “falhar e falhar” e “conseguir e conseguir”.

(Se houver uma correlação positiva, digo, então “ter sucesso e falhar” será menos provável que “conseguir e conseguir”).

Aqui, nós já temos a essência do TLC: adição sob independência leva a probabilidade a empilhar-se no meio e afilar as caudas.

Para $ n = 4 $ e $ n = 8 $, também há um pico no valor “médio” (meio termo entre o valor mínimo e valor máximo).

A intuição é a mesma — há caminhos mais simples para se obter esses resultados médios.

Se nós continuarmos, a curva sinoidal se torna mais clara.

Estamos testemunhando a [aproximação binomial da distribuição normal](https://en.wikipedia.org/wiki/De_Moivre%E2%80%93Laplace_theorem).

### Simulação 1

Desde que o TLC pareça ser mágico, simulações na rotina que verifiquem suas implicações são um bom caminho para se construir uma intuição.

Para terminar, agora tomemos a seguinte situação:

1. Escolha arbitrariamente uma distribuição $ F $ pelas observações subjacentes $ X_i $;  
2. Gere desenhos independentes de $ Y_n := \sqrt{n} ( \bar X_n - \mu ) $;  
3. Use esses desenhos para computar alguma medida dessa distribuição — como um histograma, e  
4. Compare o recente ao $ N(0, \sigma^2) $.  


Aqui está um código que faz exattamente isso para a distribuição exponencial $ F(x) = 1 - e^{- \lambda x} $.

(experimente escolher outras opções de $ F $, mas se lembre que, em conformidade com as condições do TLC, a distribuição deve ter um segundo momento finito).

```julia
using StatsPlots

function simulation1(distribution, n = 250, k = 10_000)
    σ = std(distribution)
    y = rand(distribution, n, k)
    y .-= mean(distribution)
    y = mean(y, dims = 1)
    y = √n * vec(y)
    density(y, label = "Empirical Distribution")
    return plot!(Normal(0, σ), linestyle = :dash, color = :black,
                 label = "Normal(0.00, $(σ^2))")
end
```

```julia
simulation1(Exponential(0.5))
```

O ajuste à densidade normal já é justo, e pode ser melhorado a se inserir `n`.

Você também pode experimentar outras especificações de $ F $.

### Simulação 2

Nossa próxima simulação será algo como a primeira, eceto pelo fato de que nós visamos encontrar a distribuição de $ Y_n := \sqrt{n} ( \bar X_n - \mu ) $ como $ n $ acrescenta.

Na simulação, nós estaremos trabalhando com variáveis aleatórias tendo $ \mu = 0 $.

Assim, quando $ n=1 $, teremos $ Y_1 = X_1 $, assim a distribuição inicial é já a distribuição da variável aleatória subjacente.

Para $ n=2 $, a distribuição de $ Y_2 $ é  $ (X_1 + X_2) / \sqrt{2} $, e assim por diante.

O que nós esperamos é, independentemente da distribuição da variável aleatória subjacente, a distribuição de $ Y_n $ será suavizada em uma curva sinoidal.

A próxima figura mostra esse processo para $ X_i \sim f $, onde $ f $ foi especificada como umo sendo uma combinação convexa de três diferentes densidades beta.

(Obter uma combinação convexa é um caminho fácil para produzir formas irregulares para $ f $).

```julia
function simulation2(distribution = Beta(2, 2), n = 5, k = 10_000)
    y = rand(distribution, k, n)
    for col in 1:n
        y[:,col] += rand([-0.5, 0.6, -1.1], k)
    end
    y = (y .- mean(distribution)) ./ std(distribution)
    y = cumsum(y, dims = 2) ./ sqrt.(1:5)' # retorna um grid de dados
end
```

```julia
ys = simulation2()
plots = [] # seria préalocado para otimizar o código
for i in 1:size(ys, 2)
    p = density(ys[:, i], linealpha = i, title = "n = $i")
    push!(plots, p)
end

plot(plots..., legend = false)
```

Como esperado, a distribuição suaviza-se numa curva sinoidal acrescida de $ n $.

Caso você queira, deixamos à vontade para investigar mais.

Se você rodar o arquivo do Julia comum ou do prompt iJulia,a figura deverá aparecer numa janela que você poderá rodar com seu mouse, dando visões difrentes da sequência de densidade.

### O caso multivariado


<a id='index-10'></a>
A lei dos grandes números e o teorema do limite central funcionam bem em configurações multidimensionais.

Para afirmar os resultados, vamos relembrar alguns fatos elementares sobre vetores aleatórios.

Um vetor aleatório $ \mathbf X $ é uma sequência de $ k $ variáveis aleatórias $ (X_1, \ldots, X_k) $.

Cada realização de $ \mathbf X $ é um elemento de $ \mathbb R^k $.

Uma coleção de vetores aleatórios $ \mathbf X_1, \ldots, \mathbf X_n $ é dita independente se, dados quaisquer $ n $ vetores $ \mathbf x_1, \ldots, \mathbf x_n $ in $ \mathbb R^k $, nós tivermos:

$$
\mathbb P\{\mathbf X_1 \leq \mathbf x_1,\ldots, \mathbf X_n \leq \mathbf x_n \}
= \mathbb P\{\mathbf X_1 \leq \mathbf x_1 \}
\times \cdots \times \mathbb P\{ \mathbf X_n \leq \mathbf x_n \}
$$

(A desigualdade do vetor $ \mathbf X \leq \mathbf x $ significa que $ X_j \leq x_j $ para $ j = 1,\ldots,k $)

Seja $ \mu_j := \mathbb E [X_j] $ para todos $ j =1,\ldots,k $.

O valor esperado $ \mathbb E [\mathbf X] $ de $ \mathbf X $ é definido pelo vetor de valor esperado:

$$
\mathbb E [\mathbf X]
:=
\left(
\begin{array}{c}
    \mathbb E [X_1] \\
    \mathbb E [X_2] \\
    \vdots \\
    \mathbb E [X_k]
\end{array}
\right)
=
\left(
\begin{array}{c}
    \mu_1 \\
    \mu_2\\
    \vdots \\
    \mu_k
\end{array}
\right)
=: \boldsymbol \mu
$$

A *matriz de variância-covariância* do vetor aleatório $ \mathbf X $ é definida como:

$$
\mathop{\mathrm{Var}}[\mathbf X]
:= \mathbb E
[ (\mathbf X - \boldsymbol \mu) (\mathbf X - \boldsymbol \mu)']
$$

Expandindo esse resultado, temos

$$
\mathop{\mathrm{Var}}[\mathbf X]
=
\left(
\begin{array}{ccc}
    \mathbb E [(X_1 - \mu_1)(X_1 - \mu_1)]
        & \cdots & \mathbb E [(X_1 - \mu_1)(X_k - \mu_k)] \\
    \mathbb E [(X_2 - \mu_2)(X_1 - \mu_1)]
        & \cdots & \mathbb E [(X_2 - \mu_2)(X_k - \mu_k)] \\
    \vdots & \vdots & \vdots \\
    \mathbb E [(X_k - \mu_k)(X_1 - \mu_1)]
        & \cdots & \mathbb E [(X_k - \mu_k)(X_k - \mu_k)] \\
\end{array}
\right)
$$

O $ j,k $-ésimo termo é a covariância escalar entre $ X_j $ e $ X_k $.

Com essa notação podemos proceder à LGN e TLC multivariados.

Seja $ \mathbf X_1, \ldots, \mathbf X_n $ uma sequência de vetores aleatórios independentes e identicamente distribuídos, cada um tomando valores como $ \mathbb R^k $

Seja $ \boldsymbol \mu $ o vetor $ \mathbb E [\mathbf X_i] $, e seja também $ \Sigma $ a matriz de variância-covariância de $ \mathbf X_i $.

Interpretando a adição e a multiplicação por escalar vetoriais, geralmente (isto é, pontualmente), seja

$$
\bar{\mathbf X}_n := \frac{1}{n} \sum_{i=1}^n \mathbf X_i
$$

Nessa configuração, a LGN nos diz que


<a id='equation-lln-asmv'></a>
$$
\mathbb P \left\{ \bar{\mathbf X}_n \to \boldsymbol \mu \text{ as } n \to \infty \right\} = 1 \tag{7}
$$

Aqui, $ \bar{\mathbf X}_n \to \boldsymbol \mu $ significa que $ \| \bar{\mathbf X}_n - \boldsymbol \mu \| \to 0 $, onde $ \| \cdot \| $ é a norma euclidiana padrão.

O TLC nos diz que, sendo $ \Sigma $ finito,


<a id='equation-lln-cltmv'></a>
$$
\sqrt{n} ( \bar{\mathbf X}_n - \boldsymbol \mu ) \stackrel { d } {\to} N(\mathbf 0, \Sigma)
\quad \text{as} \quad
n \to \infty \tag{8}
$$

### Exercício 1

Uma consequência bastante útil do teorema do limite central é a seguinte:

Assuma as condições do TLC como [afirmadas acima](#statement-clt).

Se $ g \colon \mathbb R \to \mathbb R $ é diferenciável em $ \mu $ and $ g'(\mu) \not= 0 $, então


<a id='equation-lln-dm'></a>
$$
\sqrt{n} \{ g(\bar X_n) - g(\mu) \}
\stackrel { d } {\to} N(0, g'(\mu)^2 \sigma^2)
\quad \text{como} \quad
n \to \infty \tag{9}
$$

Esse teorema é usado frequentemente em Estatística para obter a distribuição assintótica dos estimadores — muitos dos quais podem ser expressos como médias simples.

(Diz-se frequentemente que esses tipos de resultados usam o “método delta”)

A prova é baseada na Expansão de Taylor de $ g $ ao redor do ponto $ \mu $.

Tomando o resultado dado, seja a distribuição $ F $ de cada $ X_i $ uniforme em $ [0, \pi / 2] $ e seja $ g(x) = \sin(x) $.

Derive a distribuição assintótica de $ \sqrt{n} \{ g(\bar X_n) - g(\mu) \} $ e ilustre a convergência no mesmo sentido do programa `illustrate_clt.jl` discutido acima.

### Exercício 2

Aqui está um resultado que é usado frequentemente no desenvolvimento de testes estatísticos, e está conectado ao teorema do limite central multivariado.

Se você estuda Econometria, irá ver repetitidamente esse resultado.

Assuma a configuração do TLC multivariado [discutido acima](#multivariate-clt), e que

1. $ \mathbf X_1, \ldots, \mathbf X_n $ é uma sequência de vetores aleatórios iid, cada um obtendo valores $ \mathbb R^k $;  
1. $ \boldsymbol \mu := \mathbb E [\mathbf X_i] $, e $ \Sigma $ é a matriz de variância-covariância de $ \mathbf X_i $, e  
1. A convergência  



<a id='equation-lln-cltmv2'></a>
$$
\sqrt{n} ( \bar{\mathbf X}_n - \boldsymbol \mu ) \stackrel { d } {\to} N(\mathbf 0, \Sigma) \tag{10}
$$

é válida.

Numa configuração estatística, frequentemente se toma o lado direito normal como **padrão**, assim os intervalos de confiânça são facilmente computados.

Essa normalização pode ser alcançada com base em três observações:

Primeiro, se $ \mathbf X $ for um vetor aleatório em $ \mathbb R^k $ e $ \mathbf A $ for constante, e $ k \times k $, então

$$
\mathop{\mathrm{Var}}[\mathbf A \mathbf X]
= \mathbf A \mathop{\mathrm{Var}}[\mathbf X] \mathbf A'
$$

Segundo, pelo [teorema do mapeamento contínuo](https://en.wikipedia.org/wiki/Continuous_mapping_theorem), se $ \mathbf Z_n \stackrel{d}{\to} \mathbf Z $ em $ \mathbb R^k $ e $ \mathbf A $ for constante, além de $ k \times k $, então

$$
\mathbf A \mathbf Z_n
\stackrel{d}{\to} \mathbf A \mathbf Z
$$

Terceiro, se $ \mathbf S $ é uma matriz simétrica positiva definida $ k \times k $, então ali existe uma matriz simétrica positiva e definida $ \mathbf Q $, chamada de [raiz quadrada inversa](https://en.wikipedia.org/wiki/Square_root_of_a_matrix) de $ \mathbf S $, tal que

$$
\mathbf Q \mathbf S\mathbf Q' = \mathbf I
$$

Aqui $ \mathbf I $ é a matriz identidade $ k \times k $.

Pondo esses termos juntos, seu primeiro exercício é mostrar que, se $ \mathbf Q $ é a raiz quadrada inversa de $ \mathbf \Sigma $, então

$$
\mathbf Z_n := \sqrt{n} \mathbf Q ( \bar{\mathbf X}_n - \boldsymbol \mu )
\stackrel{d}{\to}
\mathbf Z \sim N(\mathbf 0, \mathbf I)
$$

Aplicando o teorema do mapeamento contínuo mais uma vez, temos que

$$
\| \mathbf Z_n \|^2
\stackrel{d}{\to}
\| \mathbf Z \|^2
$$

Dada a distribuição de $ \mathbf Z $, concluímos que


<a id='equation-lln-ctc'></a>
$$
n \| \mathbf Q ( \bar{\mathbf X}_n - \boldsymbol \mu ) \|^2
\stackrel{d}{\to}
\chi^2(k) \tag{11}
$$

onde $ \chi^2(k) $ é a distribuição qui-quadrado com $ k $ graus de liberdade.

(Relembre-se que $ k $ é a dimensão de $ \mathbf X_i $, os vetores aleatórios subjacentes)

Seu segundo exercício é ilustrar a convergência em [(11)](#equation-lln-ctc) com uma simulação.

Ao fazê-lo, seja

$$
\mathbf X_i
:=
\left(
\begin{array}{c}
    W_i \\
    U_i + W_i
\end{array}
\right)
$$

onde

- cada $ W_i $ é um desenho iid da distribuição uniforme em $ [-1, 1] $;  
- cada $ U_i $ é um desenho iid da distribuição uniforme em $ [-2, 2] $, e  
- $ U_i $ e $ W_i $ são independentes um do outro.  


Dicas:

1. `sqrt(A::AbstractMatrix{<:Number})` computa a raiz quadrada de `A` (você ainda deverá inverter);  
2. Você deve ser capaz de exercitar $ \Sigma $ da informação procedente.  

## Soluções

### Exercício 1

Aqui está uma solução.

```julia
function exercise1(distribution = Uniform(0, π/2); n = 250, k = 10_000, g = sin, g′ = cos)
    μ, σ = mean(distribution), std(distribution)
    y = rand(distribution, n, k)
    y = mean(y, dims = 1)
    y = vec(y)
    error_obs = sqrt(n) .* (g.(y) .- g.(μ))
    density(error_obs, label = "Empirical Density")
    return plot!(Normal(0, g′(μ) .* σ), linestyle = :dash, label = "Asymptotic",
                 color = :black)
end
exercise1()
```

O que ocorre ser você substituir $ [0, \pi / 2] $ com $ [0, \pi] $?

Nesse caso, a média $ \mu $ dessa distribuição é $ \pi/2 $, e desde que $ g' = \cos $, nós temos $ g'(\mu) = 0 $.

Assim, considerando as condições do delta, o teorema não é satisfeito.

### Exercício 2

Primeiro, verificamos o pedido de que:

$$
\sqrt{n} \mathbf Q ( \bar{\mathbf X}_n - \boldsymbol \mu )
\stackrel{d}{\to}
N(\mathbf 0, \mathbf I)
$$

Isto é, considerando os fatos apresentados no exercício.

Seja

$$
\mathbf Y_n := \sqrt{n} ( \bar{\mathbf X}_n - \boldsymbol \mu )
\quad \text{and} \quad
\mathbf Y \sim N(\mathbf 0, \Sigma)
$$

Pelo TLC multivariado e o teorema do mapeamento contínuo, nós temos

$$
\mathbf Q \mathbf Y_n
\stackrel{d}{\to}
\mathbf Q \mathbf Y
$$

Desde que as combinações lineares das variáveis aleatórias normais sejam normais, o vetor $ \mathbf Q \mathbf Y $ é também, normal.

A média é claramente $ \mathbf 0 $, e sua matriz de variância-covariância dada por

$$
\mathrm{Var}[\mathbf Q \mathbf Y]
= \mathbf Q \mathrm{Var}[\mathbf Y] \mathbf Q'
= \mathbf Q \Sigma \mathbf Q'
= \mathbf I
$$

Em conclusão,
$ \mathbf Q \mathbf Y_n \stackrel{d}{\to} \mathbf Q \mathbf Y \sim N(\mathbf 0, \mathbf I) $, o que é aquilo no qual tentávamos mostrar.

Agora vamos ao exercício de simulação.

Nossa solução é a seguinte:

```julia
function exercise2(;n = 250, k = 50_000, dw = Uniform(-1, 1), du = Uniform(-2, 2))
    vw = var(dw)
    vu = var(du)
    Σ = [vw    vw
         vw vw + vu]
    Q = inv(sqrt(Σ))
    function generate_data(dw, du, n)
        dw = rand(dw, n)
        X = [dw dw + rand(du, n)]
        return sqrt(n) * mean(X, dims = 1)
    end
    X = mapreduce(x -> generate_data(dw, du, n), vcat, 1:k)
    X = Q * X'
    X = sum(abs2, X, dims = 1)
    X = vec(X)
    density(X, label = "", xlim = (0, 10))
    return plot!(Chisq(2), color = :black, linestyle = :dash,
                 label = "Chi-squared with 2 degrees of freedom", grid = false)
end
exercise2()
```
