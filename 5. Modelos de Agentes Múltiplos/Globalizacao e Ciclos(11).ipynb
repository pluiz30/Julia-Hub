## Resumo

Nessa aula, revisaremos o artigo [Globalization and Synchronization of Innovation Cycles](http://www.centreformacroeconomics.ac.uk/Discussion-Papers/2015/CFMDP2015-27-Paper.pdf) por [Kiminori Matsuyama](http://faculty.wcas.northwestern.edu/~kmatsu/), [Laura Gardini](http://www.mdef.it/index.php?id=32) e [Iryna Sushko](http://irynasushko.altervista.org/).

Esse modelo nos ajuda a entender vários fatos estilizados interessantes sobre a economia mundial.

Um desses é ciclos de negócios sincronizados em diferentes países.

A maioria dos modelos existentes que geram ciclos de negócios sincronizados o fazem por suposição, pois vinculam a produção de cada país a um choque comum.

Eles também falham em explicar certas características dos dados, como o fato de que o grau de sincronização tende a aumentar com os vínculos comerciais.

Por outro lado, no modelo que consideramos nesta palestra, a sincronização é endógena e aumenta com a extensão da integração comercial.

Em particular, à medida que os custos comerciais caem e a concorrência internacional aumenta, os incentivos à inovação se alinham e os países sincronizam seus ciclos de inovação.

### Base

O modelo baseia-se no trabalho de Judd [[Jud85]](https://julia.quantecon.org/../zreferences.html#judd1985), Deneckner e Judd [[DJ92]](https://julia.quantecon.org/../zreferences.html#deneckere1992) e Helpman e Krugman [[HK85]](https://julia.quantecon.org/../zreferences.html#helpman1985) desenvolvendo um modelo de dois países com comércio e inovação.

No lado técnico, o artigo apresenta o conceito de [osciladores acoplados](https://en.wikipedia.org/wiki/Oscillation#Coupled_oscillations) para a modelagem econômica.

Como veremos, os osciladores acoplados surgem endogenamente dentro do modelo.

Abaixo, revisamos o modelo e replicamos alguns dos resultados da sincronização da inovação entre os países.

## Ideias Chave

É útil começar com uma visão geral do mecanismo.

### Ciclos de Inovação

Como discutido acima, dois países produzem e comercializam entre si.

Em cada país, as empresas inovam, produzindo novas variedades de bens e, ao fazê-lo, recebendo poder temporário de monopólio.

Os imitadores seguem e, após um período de monopólio, o que antes eram novas variedades agora entra em produção competitiva.

As empresas têm incentivos para inovar e produzir novos produtos quando a massa de variedades de produtos atualmente em produção é relativamente baixa.

Além disso, existem complementaridades estratégicas no momento da inovação.

As empresas têm incentivos para inovar no mesmo período, a fim de evitar competir com substitutos produzidos de forma competitiva.

Isso leva ao agrupamento temporal de inovações em cada país.

Após uma explosão de inovação, a massa de mercadorias atualmente em produção aumenta.

No entanto, os bens também se tornam obsoletos, de modo que nem todos sobrevivem de período para período.

Esse mecanismo gera um ciclo, em que a massa de variedades aumenta através da inovação simultânea e depois cai pela obsolescência.

### Sincronização

Na ausência de comércio, o momento dos ciclos de inovação em cada país é dissociado.

Este será o caso quando os custos comerciais forem proibitivamente altos.

Se os custos comerciais caem, os produtos produzidos em cada país penetram no mercado um do outro.

Como ilustrado abaixo, isso leva à sincronização dos ciclos de negócios nos dois países.

## O Modelo

Vamos escrever o modelo mais formalmente.

(O tratamento é relativamente breve, pois todos os detalhes podem ser encontrados no [artigo original](http://www.centreformacroeconomics.ac.uk/Discussion-Papers/2015/CFMDP2015-27-Paper.pdf))

O tempo é discreto com $ t = 0, 1, \dots $.

Existem dois paises indexados por $ j $ ou $ k $.

Em cada país, uma família representativa fornece inelasticamente $ L_j $ unidade de trabalho à taxa de salários de $ w_{j, t} $.

Sem perda de generalidade, assume-se que $ L_{1} \geq L_{2} $.

As famílias consomem um único bem final não comercializável, produzido de forma competitiva.

Sua produção envolve a combinação de dois tipos de insumos intermediários negociáveis via

$$
Y_{k, t} = C_{k, t} = \left( \frac{X^o_{k, t}}{1 - \alpha} \right)^{1-\alpha} \left( \frac{X_{k, t}}{\alpha} \right)^{\alpha}
$$

Aqui $ X^o_{k, t} $ é um insumo homogêneo que pode ser produzido a partir da mão de obra usando uma tecnologia linear, um por um.

É livremente negociável, fornecido competitivamente e homogêneo em todos os países.

Ao escolher o preço desse bem ser numerável e assumindo ambos os países achar que é ótimo produzir para sempre o bem homogêneo, podemos definir $ w_{1, t} = w_{2, t} = 1 $.

O bem $ X_{k, t} $ é uma composição, construído a partir de muitos bens diferenciados via

$$
X_{k, t}^{1 - \frac{1}{\sigma}} = \int_{\Omega_t} \left[ x_{k, t}(\nu) \right]^{1 - \frac{1}{\sigma}} d \nu
$$

Aqui $ x_{k, t}(\nu) $ é a quantidade total de um bem diferenciado $ \nu \in \Omega_t $ que é produzido.

O parâmetro $ \sigma > 1 $ é a elasticidade parcial direta da substituição entre um par de variedades e $ \Omega_t $ é o conjunto de variedades em cada período
$ t $.

Podemos dividir as variedades entre as que são fornecidas competitivamente e as fornecidas monopolisticamente; isto é, $ \Omega_t = \Omega_t^c + \Omega_t^m $.

### Preços

A demanda por insumos diferenciados é:

$$
x_{k, t}(\nu) = \left( \frac{p_{k, t}(\nu)}{P_{k, t}} \right)^{-\sigma} \frac{\alpha  L_k}{P_{k, t}}
$$

Aqui

- $ p_{k, t}(\nu) $ é o preço da variedade $ \nu $ e 
- $ P_{k, t} $ é o índice de preços para insumos diferenciados em $ k $,
  definido por:


$$
\left[ P_{k, t} \right]^{1 - \sigma} = \int_{\Omega_t} [p_{k, t}(\nu) ]^{1-\sigma} d\nu
$$

TO preço de uma variedade também depende da origem, $ j $, e destino, $ k $, porque as variedades de remessas entre países incorrem em um *custo de transporte de iceberg*
$ \tau_{j,k} $.

Assim, o preço efetivo no país $ k $ de uma variedade $ \nu $ produzida no país $ j $ se torna $ p_{k, t}(\nu) = \tau_{j,k} \, p_{j, t}(\nu) $.

Usando essas expressões, podemos derivar a demanda total para cada variedade,
que é:

$$
D_{j, t}(\nu) = \sum_k \tau_{j, k} x_{k, t}(\nu) = \alpha A_{j, t}(p_{j, t}(\nu))^{-\sigma}
$$

onde

$$
A_{j, t} := \sum_k \frac{\rho_{j, k}  L_{k}}{(P_{k, t})^{1 - \sigma}}
\quad \text{e} \quad
\rho_{j, k} = (\tau_{j, k})^{1 - \sigma} \leq 1
$$

É assumido que $ \tau_{1,1} = \tau_{2,2} = 1 $ e $ \tau_{1,2} = \tau_{2,1} = \tau $ para qualquer $ \tau > 1 $, de modo que:

$$
\rho_{1,2} = \rho_{2,1} = \rho := \tau^{1 - \sigma} < 1
$$

O valor $ \rho \in [0, 1) $ é uma *proxy* para o grau de globalização.

A produção de uma unidade de cada variedade diferenciada requer $ \psi $ unidades de trabalho, então o custo marginal é igual a $ \psi $ para $ \nu \in \Omega_{j, t} $.

Adicionalmente, todas as variedades competitivas terão o mesmo preço (por causa do custo marginal igual), o que significa que, para todos $ \nu \in \Omega^c $,

$$
p_{j, t}(\nu) = p_{j, t}^c := \psi
\quad \text{e} \quad
D_{j, t} = y_{j, t}^c := \alpha A_{j, t} (p_{j, t}^c)^{-\sigma}
$$

Monopolistas terão o mesmo preço de marked-up, assim, para todos $ \nu \in \Omega^m $ ,

$$
p_{j, t}(\nu) = p_{j, t}^m := \frac{\psi }{1 - \frac{1}{\sigma}}
\quad \text{and} \quad
D_{j, t} = y_{j, t}^m  := \alpha A_{j, t} (p_{j, t}^m)^{-\sigma}
$$

Defina

$$
\theta
:= \frac{p_{j, t}^c}{p_{j, t}^m} \frac{y_{j, t}^c}{y_{j, t}^m}
= \left(1 - \frac{1}{\sigma} \right)^{1-\sigma}
$$

Usando as definições anteriores e um pouco de álgebra, os índices de preços agora podem ser reescritos como:

$$
\left(\frac{P_{k,t}}{\psi}\right)^{1-\sigma} = M_{k,t}  + \rho M_{j,t}
\quad \text{onde} \quad
M_{j,t} := N_{j,t}^c + \frac{N_{j,t}^m}{ \theta}
$$

Os símbolos $ N_{j, t}^c $ e $ N_{j, t}^m $ denotarão as medidas de $ \Omega^c $ e $ \Omega^m $ respectivamente.

### Novas Variedades

Para introduzir uma nova variedade, a firma deve contratar $ f $ unidades de trabalho por variedade em cada país.

 Os lucros monopolistas devem ser menores ou iguais a zero na expectativa:

$$
N_{j,t}^m \geq 0, \quad
\pi_{j, t}^m := (p_{j, t}^m - \psi) y_{j, t}^m - f \leq 0
\quad \text{e} \quad
\pi_{j, t}^m N_{j,t}^m = 0
$$

Com outras manipulações, isso se torna:

$$
N_{j,t}^m = \theta(M_{j,t} - N_{j,t}^c) \geq 0,
\quad
\frac{1}{\sigma}
\left[
    \frac{\alpha L_j}{\theta(M_{j,t} + \rho M_{k,t})}
    +
    \frac{\alpha L_k}{\theta(M_{j,t} + M_{k,t} / \rho)}
\right]
\leq f
$$

### Lei de Movimento

Com $ \delta $ como a probabilidade exógena de uma variedade se tornar obsoleta, a equação dinâmica para a medida das firmas se torna:

$$
N_{j, t+1}^c = \delta (N_{j, t}^c + N_{j, t}^m) = \delta (N_{j, t}^c + \theta(M_{j, t} - N_{j, t}^c))
$$

Trabalharemos com uma medida normalizada de variedades:

$$
n_{j, t} := \frac{\theta \sigma f N_{j, t}^c}{\alpha (L_1 + L_2)},
\quad
i_{j, t} := \frac{\theta \sigma f N_{j, t}^m}{\alpha (L_1 + L_2)},
\quad
m_{j, t} := \frac{\theta \sigma f M_{j, t}}{\alpha (L_1 + L_2)} = n_{j, t} + \frac{i_{j, t}}{\theta}
$$

Também usamos $ s_j := \frac{L_j}{L_1 + L_2} $ a ser parcela do trabalho empregado no país $ j $.

Podemos usar essas definições e as expressões anteriores para obter uma lei de movimento para $ n_t := (n_{1, t}, n_{2, t}) $.

Em particular, dado uma condição inicial, $ n_0 = (n_{1, 0}, n_{2, 0}) \in \mathbb{R}_{+}^{2} $, a trajetória de equilíbrio, $ \{ n_t \}_{t=0}^{\infty} = \{ (n_{1, t}, n_{2, t}) \}_{t=0}^{\infty} $, é obtido por uma iteração em $ n_{t+1} = F(n_t) $ onde $ F : \mathbb{R}_{+}^{2} \rightarrow \mathbb{R}_{+}^{2} $ é dado por:

$$
\begin{aligned}
  F(n_t)
  &=
  \begin{cases}
      \big( \delta (\theta s_1(\rho) + (1-\theta) n_{1, t}), \delta (\theta s_2(\rho) + (1-\theta) n_{2, t}) \big) \; & \text{para } n_t \in D_{LL} \\
      \big( \delta n_{1, t}, \delta n_{2, t} \big) \; &\text{para } n_t \in D_{HH}  \\
      \big( \delta n_{1, t}, \delta (\theta h_2(n_{1, t}) + (1-\theta) n_{2, t}) \big) &\text{para } n_t \in D_{HL}  \\
      \big( \delta (\theta h_1(n_{2, t}) + (1-\theta) n_{1, t}, \delta n_{2, t}) \big) &\text{para } n_t \in D_{LH}
  \end{cases}
\end{aligned}
$$

Aqui

$$
\begin{aligned}
      D_{LL} & := \{ (n_1, n_2) \in \mathbb{R}_{+}^{2} | n_j \leq s_j(\rho) \} \\
      D_{HH} & := \{ (n_1, n_2) \in \mathbb{R}_{+}^{2} | n_j \geq h_j(\rho) \} \\
      D_{HL} & :=  \{ (n_1, n_2) \in \mathbb{R}_{+}^{2} | n_1 \geq s_1(\rho) \text{ e } n_2 \leq h_2(n_1) \} \\
      D_{LH} & :=  \{ (n_1, n_2) \in \mathbb{R}_{+}^{2} | n_1 \leq h_1(n_2) \text{ e } n_2 \geq s_2(\rho) \}
\end{aligned}
$$

enquanto

$$
s_1(\rho) = 1 - s_2(\rho)
= \min \left\{ \frac{s_1 - \rho s_2}{1 - \rho}, 1 \right\}
$$

e $ h_j(n_k) $ é definido implicitamente pela equação

$$
1 = \frac{s_j}{h_j(n_k) + \rho n_k} + \frac{s_k}{h_j(n_k) + n_k / \rho}
$$

Reescrever a equação acima nos dá uma equação quadrática em termos de  $ h_j(n_k) $.

Sabendo que $ h_j(n_k) > 0 $ então podemos apenas resolver a equação quadrática e retornar a raiz positiva.

Isso nos dá

$$
h_j(n_k)^2 + \left( (\rho + \frac{1}{\rho}) n_k - s_j - s_k \right) h_j(n_k) + (n_k^2 - \frac{s_j n_k}{\rho} - s_k n_k \rho) = 0
$$

## Simulação

Como veremos, isso depende das condições iniciais.

Para algumas parametrizações, a sincronização ocorrerá para a “maioria” das condições iniciais, enquanto para outras a sincronização será rara.

```julia
function h_j(j, nk, s1, s2, θ, δ, ρ)
    # Descubra quem é h que estamos avaliando 
    if j == 1
        sj = s1
        sk = s2
    else
        sj = s2
        sk = s1
    end

    # Coeficienciete no eixo quadrático a x^2 + b x + c = 0
    a = 1.0
    b = ((ρ + 1 / ρ) * nk - sj - sk)
    c = (nk * nk - (sj * nk) / ρ - sk * ρ * nk)

    # Solução positiva da forma quadrática
    root = (-b + sqrt(b * b - 4 * a * c)) / (2 * a)

    return root
end

DLL(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ) =
    (n1 ≤ s1_ρ) && (n2 ≤ s2_ρ)

DHH(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ) =
    (n1 ≥ h_j(1, n2, s1, s2, θ, δ, ρ)) && (n2 ≥ h_j(2, n1, s1, s2, θ, δ, ρ))

DHL(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ) =
    (n1 ≥ s1_ρ) && (n2 ≤ h_j(2, n1, s1, s2, θ, δ, ρ))

DLH(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ) =
    (n1 ≤ h_j(1, n2, s1, s2, θ, δ, ρ)) && (n2 ≥ s2_ρ)

function one_step(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)
    # Dependendo de onde estamos, avalie o ramo direito 
    if DLL(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)
        n1_tp1 = δ * (θ * s1_ρ + (1 - θ) * n1)
        n2_tp1 = δ * (θ * s2_ρ + (1 - θ) * n2)
    elseif DHH(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)
        n1_tp1 = δ * n1
        n2_tp1 = δ * n2
    elseif DHL(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)
        n1_tp1 = δ * n1
        n2_tp1 = δ * (θ * h_j(2, n1, s1, s2, θ, δ, ρ) + (1 - θ) * n2)
    elseif DLH(n1, n2, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)
        n1_tp1 = δ * (θ * h_j(1, n2, s1, s2, θ, δ, ρ) + (1 - θ) * n1)
        n2_tp1 = δ * n2
    end

    return n1_tp1, n2_tp1
end

new_n1n2(n1_0, n2_0, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ) =
    one_step(n1_0, n2_0, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)

function pers_till_sync(n1_0, n2_0, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ,
                        maxiter, npers)

    # Inicialize o status da sincronização 
    synchronized = false
    pers_2_sync = maxiter
    iters = 0

    nsync = 0

    while (~synchronized) && (iters < maxiter)
        # Incrementa o número de iterações e obtém os próximos valores 
        iters += 1

        n1_t, n2_t = new_n1n2(n1_0, n2_0, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)

        # Verifique se o mesmo nesse período 
        if abs(n1_t - n2_t) < 1e-8
            nsync += 1
        # Se não, então reinicie o controlador nsync 
        else
            nsync = 0
        end

        # Se temos sido em sincronia para npers em seguida, parar e países 
        #  tornou-se sincronizado períodos Nsync atrás 
        if nsync > npers
            synchronized = true
            pers_2_sync = iters - nsync
        end
        n1_0, n2_0 = n1_t, n2_t
    end
    return synchronized, pers_2_sync
end

function create_attraction_basis(s1_ρ, s2_ρ, s1, s2, θ, δ, ρ,
                                 maxiter, npers, npts)
    # Criar intervalo de unidades com npts
    synchronized, pers_2_sync = false, 0
    unit_range = range(0.0,  1.0, length = npts)

    # Aloca espaço para armazenar tempo para sync
    time_2_sync = zeros(npts, npts)
    # itera sobre as condições iniciais
    for (i, n1_0) in enumerate(unit_range)
        for (j, n2_0) in enumerate(unit_range)
            synchronized, pers_2_sync = pers_till_sync(n1_0, n2_0, s1_ρ, s2_ρ,
                                                       s1, s2, θ, δ, ρ,
                                                       maxiter, npers)
            time_2_sync[i, j] = pers_2_sync
        end
    end
    return time_2_sync
end

# o modelo
function MSGSync(s1 = 0.5, θ = 2.5, δ = 0.7, ρ = 0.2)
    # Armazene outros pontos de corte e parâmetros que usamos
    s2 = 1 - s1
    s1_ρ = min((s1 - ρ * s2) / (1 - ρ), 1)
    s2_ρ = 1 - s1_ρ

    return (s1 = s1, s2 = s2, s1_ρ = s1_ρ, s2_ρ = s2_ρ, θ = θ, δ = δ, ρ = ρ)
end

function simulate_n(model, n1_0, n2_0, T)
    # Unpack parâmetros
    @unpack s1, s2, θ, δ, ρ, s1_ρ, s2_ρ = model

    # Aloca espaço
    n1 = zeros(T)
    n2 = zeros(T)

    # Simula para T periodos
    for t in 1:T
        # Obtém os novos valores
        n1[t], n2[t] = n1_0, n2_0
        n1_0, n2_0 = new_n1n2(n1_0, n2_0, s1_ρ, s2_ρ, s1, s2, θ, δ, ρ)
    end

    return n1, n2
end

function pers_till_sync(model, n1_0, n2_0,
                        maxiter = 500, npers = 3)
    # Unpack parâmetros
    @unpack s1, s2, θ, δ, ρ, s1_ρ, s2_ρ = model

    return pers_till_sync(n1_0, n2_0, s1_ρ, s2_ρ, s1, s2,
                        θ, δ, ρ, maxiter, npers)
end

function create_attraction_basis(model;
                                 maxiter = 250,
                                 npers = 3,
                                 npts = 50)
    # Unpack parâmetros
    @unpack s1, s2, θ, δ, ρ, s1_ρ, s2_ρ = model

    ab = create_attraction_basis(s1_ρ, s2_ρ, s1, s2, θ, δ,
                                 ρ, maxiter, npers, npts)
    return ab
end
```

### Séries Temporais das Medidas das Firmas 

Escrevemos abaixo uma função curta que explora o código anterior e plota duas séries temporais.

Cada série cronológica fornece a dinâmica para os dois países.

As séries temporais compartilham parâmetros, mas diferem em sua condição inicial.

Aqui está a função:

```julia
function plot_timeseries(n1_0, n2_0, s1 = 0.5, θ = 2.5, δ = 0.7, ρ = 0.2)
    model = MSGSync(s1, θ, δ, ρ)
    n1, n2 = simulate_n(model, n1_0, n2_0, 25)
    return [n1 n2]
end

# Cria figuras
data_ns = plot_timeseries(0.15, 0.35)
data_s = plot_timeseries(0.4, 0.3)

plot(data_ns, title = "Not Synchronized", legend = false)
```

```julia
plot(data_s, title = "Synchronized", legend = false)
```

No primeiro caso, a inovação nos dois paises não sincroniza.

No segundo caso condições iniciais diferentes são escolhidas, e os ciclos se tornam sincronizados.

```julia
function plot_attraction_basis(s1 = 0.5, θ = 2.5, δ = 0.7, ρ = 0.2; npts = 250)
    # Crie a base de atração
    unitrange = range(0,  1, length = npts)
    model = MSGSync(s1, θ, δ, ρ)
    ab = create_attraction_basis(model,npts=npts)
    plt = Plots.heatmap(ab, legend = false)
end
```

```julia
params = [[0.5, 2.5, 0.7, 0.2],
          [0.5, 2.5, 0.7, 0.4],
          [0.5, 2.5, 0.7, 0.6],
          [0.5, 2.5, 0.7, 0.8]]

plots = (plot_attraction_basis(p...) for p in params)
plot(plots..., size = (1000, 1000))
```
